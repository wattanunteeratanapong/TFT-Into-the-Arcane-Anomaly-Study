```python
import random
import matplotlib.pyplot as plt
```

```python
target_anomaly = 1
simulations = 10000000
stat = [] 
```

```python
for _ in range(simulations):  
    trial = 1
    anomaly_pool = list(range(1, 61))  
    
    while anomaly_pool:  
        random_anomaly = random.choice(anomaly_pool)  
        anomaly_pool.remove(random_anomaly)  
        
        if random_anomaly == target_anomaly:
            stat.append(trial)
            break
        else:
            trial += 1
```

```python
plt.hist(stat, bins=range(1, max(stat) + 2), align='left', rwidth=0.8, density=True)
plt.xlabel('Trials')
plt.ylabel('Probability')
plt.title('Distribution of Trials to Found Target Anomaly')
plt.show()
```
